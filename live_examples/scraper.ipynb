### Modules

```python
import requests
from lxml import html
```

### Download page

```python
link = 'https://www.beeradvocate.com/beer/profile/147/38470/?view=beer&sort=topr&start=0'
page = requests.get(link)
print(page)
print(page.text)
```

### Parse Page

```python
tree = html.fromstring(page.content)
```

### Navigate Tree

```python
p = '//*[@id="rating_fullview_content_2"]'
text_selector = '//text()'
tree.xpath(p + text_selector)
```

### Get Review

```python
reviews = tree.xpath('//div[@id="rating_fullview_content_2"][1]/text()')
reviews = ' '.join(reviews)
print(reviews)
```

### Clean Reviews

```python
reviews = reviews.split('\xa0\xa0rDev')
reviews = [r for r in reviews if r]
reviews = [r.replace('\xa0', '') for r in reviews]
reviews = [r.replace('\n', '') for r in reviews]
print(reviews)
```

### Get Ratings

```python
ratings = tree.xpath('//*[@id="rating_fullview_content_2"]/span[1]/text()')
ratings = [float(r) for r in ratings]
print(ratings)
```
